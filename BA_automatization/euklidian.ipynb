```python
# Notebook to calculate the euklidian distance between sampled Nodes out of one Embedding file
# after calculating the distance, compare the mean between different embedding files

import pykeen
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import math
import scipy
from scipy.spatial import distance
from scipy.spatial import distance_matrix
```

```python
# function to locate the nodes from fitst data in second data,
# builds two matrizes and calculates the distance matrix for both data sets
# takes in list of dataFrames 
def calculate_difference(data_list, sample_size):
    
    # sample out of first list:
    sample= data_list[0].sample(sample_size)
    # store the sampled nodes as string in list
    sampled_nodes = list(sample.nodes)
    
    # store distance matrices
    means=[]
    # for loop to set index, stack the embeddings and calculate the normalization
    for dataF in data_list:
        dataF.set_index('nodes', drop=True, inplace=True)
        #accsess the Embedding-array in the pd.Series-array
        d_list =np.stack(dataF.loc[sampled_nodes].embeddings.values)
        #calculate the mean
        d_list=d_list - np.mean(d_list)
        d_list=d_list / np.std(d_list)
        
        # calculate the distance matrix and the mean 
        distance_m = distance_matrix(d_list, d_list, p=2)
        means.append(np.mean(distance_m))
        
    
    # return the mean list 
    return means
```
